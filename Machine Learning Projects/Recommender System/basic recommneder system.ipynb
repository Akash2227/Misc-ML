```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('tmdb_5000_credits.csv')
df1.head()
```

```python
df2 = pd.read_csv('tmdb_5000_movies.csv')
df2.head()
```

```python
print('credits',df1.shape)
print('movies',df2.shape)
```

```python
df_credits = df1.rename(index = str,columns = {'movie_id':'id'})
df_cm = df2.merge(df_credits,on = 'id')
```

```python
df_cm.head()
```

```python
df_cm.shape
```

```python
df_cmf = df_cm.drop(['homepage','title_x','title_y','production_countries'],axis = 'columns')
```

```python
df_cmf.head()
```

```python
df_cmf.info()
```

```python
v = df_cmf['vote_count']
r = df_cmf['vote_average']
c = df_cmf['vote_average'].mean()
m = df_cmf['vote_count'].quantile(0.70)
```

```python
df_cmf['weighted_Average'] = (r*v + c*m)/(v + m)
```

```python
df_cmf
```

```python
df_cmfs = df_cmf.sort_values(['weighted_Average'],ascending = False)
```

```python
df_cmfs1=df_cmfs[['original_title','vote_average','vote_count','weighted_Average','popularity']]
```

```python
import seaborn as sn
%matplotlib inline
```

```python
weight_average = df_cmfs1.sort_values(['weighted_Average'],ascending = False)
plt.figure(figsize = (12,6))
axis1 = sn.barplot(x = weight_average['weighted_Average'].head(10),y = weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.title('best movies by weighted Average')
```

```python
popularity = df_cmfs1.sort_values(['popularity'],ascending = False)
```

```python
popularity.head()
```

```python
plt.figure(figsize = (12,6))
axis2 = sn.barplot(x = popularity['popularity'].head(),y = popularity['original_title'].head())
plt.title('Based on popularity of the movie')
```
